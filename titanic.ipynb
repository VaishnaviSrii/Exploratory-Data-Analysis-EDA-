```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel("C:/Users/ACER/Desktop/DataAnalyticsTasks/titanicDataset.xlsx")
```

```python
print(data)
```

```python
data.describe()
data.info()
data.value_counts()
data.isnull().sum()
```

```python
#plotting histograms,boxplots, scatterplots
#histograms

data.hist(figsize=(14,10))
plt.tight_layout()
plt.show
```

```python
print(data.columns)
```

```python
#boxplot
sns.boxplot(x='Age', data=data)
plt.show()
```

```python
#pairplots
sns.pairplot(data)
plt.show()
```

```python
print(data.dtypes)
```

```python
numeric_data = data.select_dtypes(include='number')
print(numeric_data.columns)
```

```python
#heatmap 
plt.figure(figsize=(8, 4))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
#scatterplots
sns.scatterplot(x='Age', y='Sex', data=data)
plt.show()
```
